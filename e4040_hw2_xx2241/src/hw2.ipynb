### Problem c: Bullet 2: Compare the two cases

### When we use dropout hidden layers, the test accuracy is 85.27%. When we use hidden layers without dropout the test accuracy is 80.71%. With all other parameter fixed up, dropout hidden layers can increase about 5% of test accuracy. That is because when we use dropout hidden layers, it will cripple neural network by removing hidden units stochastically, which will increase the co-adaptation of remaining hidden units with others. This will finally lead to the increase of test accuracy.
