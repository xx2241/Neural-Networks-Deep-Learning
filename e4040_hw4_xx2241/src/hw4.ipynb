```python
from hw4a import test_rnnslu
emb, idx2label, idx2word = test_rnnslu(**{
        'fold': 3,
        'lr': 0.0970806646812754,
        'verbose': False,
        'decay': True,
        'win': 7,
        'nhidden': 200,
        'seed': 345,
        'emb_dimension': 50,
        'nepochs': 60,
        'savemodel': False,
        'normal': True,
        'folder':'../result'})
```

```python
from scipy import spatial
import numpy as np
emb = np.asarray(emb)


words = {}
for i, v in idx2word.iteritems():
    if v == 'to':
        words[v] = emb[i]
    if 'UNK' not in v and 'DIGIT' not in v and "'" not in v and len(v) > 4:
        words[v] = emb[i]

word_dict = {}
for w1 in words.keys():
    for w2 in words.keys():
        k = w1 + '-' + w2
        if w1 != w2 and k not in word_dict:
            word_dict[k] = words[w1] - words[w2]
            


print(spatial.distance.cosine(word_dict['sunday-people'], word_dict['people-costs']))
print(spatial.distance.cosine(word_dict['flying-seattle'], word_dict['flying-minneapolis']))
print(spatial.distance.cosine(word_dict['expensive-costs'], word_dict['maximum-costs']))
print(spatial.distance.cosine(word_dict['sunday-requesting'], word_dict['friday-requesting']))
print(spatial.distance.cosine(word_dict['sunday-requesting'], word_dict['friday-requesting']))
print(spatial.distance.cosine(word_dict['daily-aircraft'], word_dict['california-aircraft']))
```

#### As we can see from above, when the word pairs are irrelevant such as "sunday/people" with "people/costs", the distance between the two vector are large. When the words paris very similar such as "sunday/requesting" with "friday/requesting", the distance between the two vector are very small. When the word pairs have some kind of relation for example "flying/seattle" with "flying/minneapolis", the distance between the two vector are relatively small.
